## Function transformers ðŸ“š
To perform function transformers and power transformers using the `sklearn` library on a toy dataset, and visualize the results, we'll go step-by-step, interactively explaining each part. Let's get started! ðŸš€

### Step 1: Import Libraries ðŸ“š
First, we'll need to import the necessary libraries, including `numpy`, `matplotlib`, and `sklearn`.

````python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.preprocessing import FunctionTransformer, PowerTransformer
from sklearn.pipeline import Pipeline
````

### Step 2: Create a Toy Dataset ðŸŽ²
We'll use `make_regression` from `sklearn` to create a simple toy dataset.

````python
# Create toy data
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)

# Visualize the raw data
plt.scatter(X, y, color='blue', label='Raw Data')
plt.title('Toy Dataset')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
````

````python
# QQ Plot for original data
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
stats.probplot(X.flatten(), dist="norm", plot=plt)
plt.title('QQ Plot of Raw Data')

# KDE Plot for original data
plt.subplot(1, 2, 2)
sns.kdeplot(X.flatten(), color='blue', shade=True)
plt.title('KDE Plot of Raw Data')
plt.xlabel('Feature')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
````

### Step 3: Apply Function Transformer ðŸ”§
A `FunctionTransformer` can be used to apply any arbitrary function to our data. We'll apply the natural logarithm to the feature data to transform it.

````python
# Apply log transformation using FunctionTransformer
log_transformer = FunctionTransformer(np.log1p, validate=True)

# Transform the data
X_log = log_transformer.transform(X)

# Visualize the transformed data
plt.scatter(X_log, y, color='green', label='Log-Transformed Data')
plt.title('Log Transformation using FunctionTransformer')
plt.xlabel('Log(Feature)')
plt.ylabel('Target')
plt.legend()
plt.show()
````

### Step 4: Apply Power Transformer âš¡
A `PowerTransformer` is a more advanced technique that stabilizes variance and makes the data more Gaussian-like. Let's apply it to our data.

````python
# Apply power transformation
power_transformer = PowerTransformer(method='yeo-johnson', standardize=False)

# Transform the data
X_power = power_transformer.fit_transform(X)

# Visualize the transformed data
plt.scatter(X_power, y, color='orange', label='Power-Transformed Data')
plt.title('Power Transformation using PowerTransformer')
plt.xlabel('Power-Transformed Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
````

### Step 5: Compare All Transformations ðŸ“Š
Now, let's visualize the original, log-transformed, and power-transformed data side by side for comparison.

````python
# Plot all transformations
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(X, y, color='blue', label='Raw Data')
plt.title('Raw Data')
plt.xlabel('Feature')
plt.ylabel('Target')

plt.subplot(1, 3, 2)
plt.scatter(X_log, y, color='green', label='Log-Transformed Data')
plt.title('Log Transformation')
plt.xlabel('Log(Feature)')

plt.subplot(1, 3, 3)
plt.scatter(X_power, y, color='orange', label='Power-Transformed Data')
plt.title('Power Transformation')
plt.xlabel('Power-Transformed Feature')

plt.tight_layout()
plt.show()
````

### Interactive Description ðŸŽ¨

1. **Raw Data**: This is our original data, generated randomly with some noise. It represents a linear relationship with some variance. ðŸ“‰

2. **Log Transformation**: Applying a logarithm using `FunctionTransformer` compresses the range of our data, particularly useful for dealing with features with large ranges or exponential growth. This makes the data less skewed. ðŸ”¢

3. **Power Transformation**: Using `PowerTransformer` (with Yeo-Johnson method), we standardize the data in a way that stabilizes variance and makes it more Gaussian-like. It's particularly helpful when data contains negative values or isn't normally distributed. âš¡

### Step 6: Import Additional Libraries ðŸ› ï¸

````python
import scipy.stats as stats
import seaborn as sns
````

### Step 7: QQ Plots and KDE Plots for Original Data ðŸŽ¯
Let's start by visualizing the distribution of the original data using QQ plots and KDE plots.

````python
# QQ Plot for original data
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
stats.probplot(X.flatten(), dist="norm", plot=plt)
plt.title('QQ Plot of Raw Data')

# KDE Plot for original data
plt.subplot(1, 2, 2)
sns.kdeplot(X.flatten(), color='blue', shade=True)
plt.title('KDE Plot of Raw Data')
plt.xlabel('Feature')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
````

````python

### Step 8: QQ Plots and KDE Plots for Log-Transformed Data ðŸ”„

Now, let's do the same for the log-transformed data.

```python
# QQ Plot for log-transformed data
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
stats.probplot(X_log.flatten(), dist="norm", plot=plt)
plt.title('QQ Plot of Log-Transformed Data')

# KDE Plot for log-transformed data
plt.subplot(1, 2, 2)
sns.kdeplot(X_log.flatten(), color='green', shade=True)
plt.title('KDE Plot of Log-Transformed Data')
plt.xlabel('Log(Feature)')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

### Step 9: QQ Plots and KDE Plots for Power-Transformed Data âš¡

Finally, let's visualize the power-transformed data.

```python
# QQ Plot for power-transformed data
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
stats.probplot(X_power.flatten(), dist="norm", plot=plt)
plt.title('QQ Plot of Power-Transformed Data')

# KDE Plot for power-transformed data
plt.subplot(1, 2, 2)
sns.kdeplot(X_power.flatten(), color='orange', shade=True)
plt.title('KDE Plot of Power-Transformed Data')
plt.xlabel('Power-Transformed Feature')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

### Interactive Description of the Plots ðŸŽ¨

1. **QQ Plot of Raw Data**: This plot compares the quantiles of our original data against a standard normal distribution. If the data is normally distributed, the points will follow a straight diagonal line. In our case, the raw data might show deviations due to noise or skewness. ðŸŒ€

2. **KDE Plot of Raw Data**: The Kernel Density Estimate plot gives us a smoothed version of the histogram, showing the probability density function of the original feature. It helps us visualize the underlying distribution of the data. ðŸ”

3. **QQ Plot of Log-Transformed Data**: After applying the log transformation, the data should appear more normally distributed if it was skewed initially. The points on the QQ plot should align closer to the diagonal line compared to the raw data. ðŸ“

4. **KDE Plot of Log-Transformed Data**: This KDE plot shows the density distribution after the log transformation. The distribution is expected to be more symmetric and less skewed. ðŸ”„

5. **QQ Plot of Power-Transformed Data**: The power-transformed data is expected to be even closer to a normal distribution, especially if the original data had significant skewness. The QQ plot should show a better alignment with the diagonal. âš¡

6. **KDE Plot of Power-Transformed Data**: This KDE plot reveals the density distribution after applying the power transformation. It should appear more Gaussian-like with reduced skewness or kurtosis. ðŸ“Š

### Conclusion ðŸ
By applying QQ plots and KDE plots, we get a clearer picture of how the transformations (log and power) affect the distribution of our data, bringing it closer to normality. This is crucial for many machine learning algorithms that assume normal distribution of the input features.
````
